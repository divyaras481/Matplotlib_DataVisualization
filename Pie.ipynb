Display percentage & change distance inside each wedge

```python
plt.pie(votes , labels=movie, explode=[0,0.1,0,0.1],startangle=0,autopct="%0.2f%%",radius=1)
plt.title('Types of movie')
plt.legend(loc=4)
plt.show()
```

Set labels size of the pie chart

```python
plt.pie(votes ,
        labels=movie,
        explode=[0,0.1,0,0.1],
        startangle=0,
        autopct="%0.2f%%",
        radius=1,
       textprops={'fontsize':12}) #we can pass required changes in this
plt.title('Types of movie')
plt.legend(loc=4)
plt.show()
```

Border of the wedges

```python
plt.pie(votes ,
        labels=movie,
        explode=[0,0.1,0,0.1],
        startangle=0,
        autopct="%0.2f%%",
        radius=1,
       textprops={'fontsize':12},
       wedgeprops={'edgecolor':'black','linewidth':1,'linestyle':':'}) 
plt.title('Types of movie')
plt.legend(loc=4)
plt.show()
```

Add a shadow to the pie chart

```python
plt.pie(votes ,
        labels=movie,
        explode=[0,0.1,0,0.1],
        startangle=0,
        autopct="%0.2f%%",
        radius=1,
       textprops={'fontsize':12},
       wedgeprops={'edgecolor':'black','linewidth':1,'linestyle':':'},
       shadow=True) 
plt.title('Types of movie')
plt.legend(loc=4)
plt.show()
```
