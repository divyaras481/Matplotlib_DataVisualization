```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
grades =['A','B','C','D','E','F']
test1 = [3,13,10,4,5,12]
```

```python
#Draw a bar plot
plt.bar(grades,test1)
plt.grid()
plt.show()
```

#give title, xlabel , ylabel

```python
plt.bar(grades,test1)
plt.title('Grades of the test')
plt.xlabel('grades')
plt.ylabel('marks')
plt.show()
```

changing font

```python
plt.bar(grades,test1)
plt.title('Grades of the test',fontsize='15')
plt.xlabel('grades')
plt.ylabel('marks')
plt.show()
```

changing color of bar,edgecolor,linestyle

```python
plt.bar(grades,test1, width=0.4 , color='y',edgecolor='red' , linestyle=':',lw=2)
plt.title('Grades of the test',fontsize='15')
plt.xlabel('grades')
plt.ylabel('marks')
plt.show()
```

```python
plt.figure(figsize=(10,3), facecolor='lightgreen',linewidth=2,edgecolor='black')
plt.bar(grades,test1, width=0.4 , color='y',edgecolor='red' , linestyle=':',lw=2)
plt.title('Grades of the test',fontsize='15')
plt.xlabel('grades')
plt.ylabel('marks')
plt.show()
```
