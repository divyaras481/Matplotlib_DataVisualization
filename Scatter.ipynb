```python
import matplotlib.pyplot as plt
```

```python
temp=[14.2,16.4,11.9,15.2,18.5,22.1,19.4]
waffle_sales=[21,32,18,33,40,52,41]
coffee_sales=[80,65,89,70,56,45,43]
```

```python
plt.scatter(temp,waffle_sales)
plt.show()
```

```python
plt.figure(figsize=(10,4),edgecolor='black',linewidth=2,facecolor='pink')
plt.scatter(temp,waffle_sales, s=50, color='green',linewidth=2,alpha=1,label='waffle_sales')
plt.scatter(temp,coffee_sales, s=50, color='red',linewidth=2,alpha=1,label='coffee_sales')

plt.title('Waffle vs Coffee')
plt.xlabel('temp')
plt.ylabel('Waffle Data')
plt.legend()
plt.show()
```
